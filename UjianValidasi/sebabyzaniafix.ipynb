```python
from bs4 import BeautifulSoup
import requests
import urllib.request
import csv
from urllib.request import urlopen
from datetime import datetime
import time
import pandas as pd
import os

with open ('F:\\SKRIPSI\\Topik2\\UjianValidasi\\LinkURL\\url_babyzania.csv') as f:
    urls=(line.strip()for line in f)
    baby_list= []
    for row in urls:
        row= row.replace('"','')
        print(row)
        r = requests.get(row)
        soup = BeautifulSoup(r.text, "html.parser")
        all_product= soup.find_all('div',{'class':"product-grid-item"})
        
        for item in all_product:
            try:
                linkEcommerce = row
            except :
                linkEcommerce = ''
            try:
                url= item.find_all('a')[0]
                linkProduk= url.get('href')
            except:
                linkProduk=''
            try:
                name= item.find_all('img')[0]
                namaProduk= name.get('title')
            except:
                namaProduk=''
            try:
                diskon= item.find_all('b')[0].text
                diskonProduk= diskon.replace('-','')
                diskonProduk= diskonProduk.replace('%','')
                diskonProduk= int(diskonProduk)/100
            except:
                diskonProduk=''
            try:
                hargaLama= item.find_all('span',{"class":"price-old"})[0].text
                hargaLama= hargaLama.replace('Rp ','')
                hargaLama= hargaLama.replace('.','')
            except:
                harga_lama=item.find_all('p', {"class":"price"})[0].text
                harga_lama= harga_lama.replace('                                    ','')
                harga_lama= harga_lama.replace('                                                    ','')
                hargaLama= harga_lama.replace('\n','')
                hargaLama= hargaLama.replace('Rp ','')
                hargaLama= hargaLama.replace('.','')
            try:
                hargaBaru= item.find_all('span',{"class":'price-new'})[0].text
                hargaBaru= hargaBaru.replace('Rp ','')
                hargaBaru= hargaBaru.replace('.','')
            except:
                harga_diskon=item.find_all('p', {"class":"price"})[0].text
                harga_diskon=harga_diskon.replace('                                    ','')
                harga_diskon=harga_diskon.replace('                                                    ','')
                hargaBaru=harga_diskon.replace('\n','')
                hargaBaru= hargaBaru.replace('Rp ','')
                hargaBaru= hargaBaru.replace('.','')
            try:
                u=urlopen(linkProduk)
                url_produk_soup= BeautifulSoup(u,'html.parser') 
                brand= url_produk_soup.find('li', {"class":"p-brand"})
                brandProduk= brand.find_all('a')[0].text
            except:
                brandProduk=''
            try:
                u=urlopen(linkProduk)
                url_produk_soup= BeautifulSoup(u,'html.parser')
                kategori=url_produk_soup.find('ul',{"class":"breadcrumb"})
                kategori=kategori.find_all('span')
                kategori=str(kategori)
                kategori= kategori.replace('<span itemprop="title">','')
                kategoriProduk= kategori.replace('</span>','')
            except:
                kategoriProduk=''
                
            babyzania=[linkEcommerce,namaProduk,linkProduk,kategoriProduk,brandProduk, hargaLama,hargaBaru,diskonProduk, datetime.now().strftime("%y-%m-%d %H:%M")]
            baby_list.append(babyzania)
            
    path = 'F:\\SKRIPSI\\Topik2\\UjianValidasi\\HasilScraper\\'
    now = datetime.now()
    filename = 'babyzania_%s_%s_%s %s_%s_%s_%s.csv'%(now.day,now.month,now.year,now.hour,now.minute,now.second,now.microsecond)
    df=pd.DataFrame(baby_list, columns=['linkEcommerce','namaProduk', 'linkProduk', 'kategoriProduk', 'brandProduk', 'hargaLama', 'hargaBaru','diskonProduk', 'tgl update'])
    df.to_csv(os.path.join(path, filename))    
            
            
```
