```python
from bs4 import BeautifulSoup
import requests
import urllib.request
import csv
from urllib.request import urlopen
from datetime import datetime
import time
import re

with open ('F:\\SKRIPSI\\Topik2\\UjianValidasi\\LinkURL\\url_klikmart.csv') as f:
    urls=(line.strip()for line in f)
    klik_list=[]
    for row in urls:
        r= urlopen(row)
        soup = BeautifulSoup(r, "html.parser")
        
        try:
            page= soup.find('div',{"class":"pagination"})
            list_page= page.find_all('option')
            total_page = len(list_page)
            
            print(total_page)
            
           
        except:
            total_page='0'
        product= soup.find("div", {"class":"box-item clearfix"})
        
        if total_page == 1 :
            print(row)
            all_product= product.find_all("div", {"class":"item"})
            for item in all_product:
                try:
                    linkEcommerce = row
                except :
                    linkEcommerce = ''
                try:
                    namaProduk = item.find_all('div',{'class':'wrp-content'})[0]
                    namaProduk = namaProduk.find_all('div',{'class':'title'})[0].text
                    namaProduk = namaProduk.replace('                            ','')
                    namaProduk = namaProduk[2:]
                except:
                    namaProduk = ''
                try:
                    linkProduk = item.find_all('a')[0]
                    linkProduk = linkProduk.get('href')
                    linkProduk = linkProduk.replace(linkProduk, 'http://www.klikindomaret.com/'+linkProduk)
                except:
                    linkProduk = ''
                try:
                    hargaBaru = item.find_all('span',{'class':'normal price-value'})[0].text
                    hargaBaru = hargaBaru.replace('Rp ','')
                    hargaBaru = hargaBaru.replace('.','')
                    
                except:
                    hargaBaru = ''
                try:
                    hargaLama = item.find_all('span',{'class':'strikeout disc-price'})[0].text
                    hargaLama = hargaLama.replace('Rp ','')
                    hargaLama = hargaLama.replace('.','')
                    diskonProduk = item.find_all('span',{'class':'discount'})[0].text
                    hargaLama = hargaLama.replace('                                                                        ','')
                    hargaLama = hargaLama.replace(diskonProduk,'')
                    hargaLama = int(hargaLama)
                except:
                    hargaLama = hargaBaru
                try:
                    diskonProduk = item.find_all('span',{'class':'discount'})[0].text
                    diskonProduk = diskonProduk.replace('%','')
                    diskonProduk = int(diskonProduk)/100
                except:
                    diskonProduk = ''
                    
                klik = [linkEcommerce, namaProduk, hargaBaru,  hargaLama, diskonProduk,linkProduk, datetime.now().strftime("%y-%m-%d %H:%M")]
                klik_list.append(klik)
                
            
        else:
            total_page = int(total_page)
            for i in range(total_page):
                change_url = row.replace(row,row+'?categories=minuman-ringan&productbrandid=&sortcol=PROMO&pagesize=30&page=%s&startprice=&endprice=&attributes='%(i+1))
                print(change_url)
                c= urlopen(change_url)
                next_soup = BeautifulSoup(c, "html.parser")
                product= next_soup.find("div", {"class":"box-item clearfix"})
                all_product= product.find_all("div", {"class":"item"})
                for item in all_product:
                    try:
                        linkEcommerce = row
                    except :
                        linkEcommerce = ''
                    try:
                        namaProduk = item.find_all('div',{'class':'wrp-content'})[0]
                        namaProduk = namaProduk.find_all('div',{'class':'title'})[0].text
                        namaProduk = namaProduk.replace('                            ','')
                        namaProduk = namaProduk[2:]
                    except:
                        namaProduk = ''
                    try:
                        linkProduk = item.find_all('a')[0]
                        linkProduk = linkProduk.get('href')
                        linkProduk = linkProduk.replace(linkProduk, 'http://www.klikindomaret.com/'+linkProduk)
                    except:
                        linkProduk = ''
                    try:
                        hargaBaru = item.find_all('span',{'class':'normal price-value'})[0].text
                        hargaBaru = hargaBaru.replace('Rp ','')
                        hargaBaru = hargaBaru.replace('.','')

                    except:
                        hargaBaru = ''
                    try:
                        hargaLama = item.find_all('span',{'class':'strikeout disc-price'})[0].text
                        hargaLama = hargaLama.replace('Rp ','')
                        hargaLama = hargaLama.replace('.','')
                        diskonProduk = item.find_all('span',{'class':'discount'})[0].text
                        hargaLama = hargaLama.replace('                                                                        ','')
                        hargaLama = hargaLama.replace(diskonProduk,'')
                        hargaLama = int(hargaLama)
                    except:
                        hargaLama = hargaBaru
                    try:
                        diskonProduk = item.find_all('span',{'class':'discount'})[0].text
                        diskonProduk = diskonProduk.replace('%','')
                        diskonProduk = int(diskonProduk)/100
                    except:
                        diskonProduk = ''

                    klik = [linkEcommerce, namaProduk, hargaBaru,  hargaLama, diskonProduk,linkProduk, datetime.now().strftime("%y-%m-%d %H:%M")]
                    klik_list.append(klik)
                
    import pandas as pd
    import os
    path = 'F:\\SKRIPSI\\Topik2\\UjianValidasi\\HasilScraper\\'
    now = datetime.now()
    filename = 'klikmart_%s_%s_%s %s_%s_%s_%s.csv'%(now.day,now.month,now.year,now.hour,now.minute,now.second,now.microsecond)
    df=pd.DataFrame(klik_list, columns=['linkEcommerce','namaProduk', 'hargaBaru',  'hargaLama', 'diskonProduk','linkProduk',  'tgl_update' ])
    df.to_csv(os.path.join(path, filename))               
                    
                
```
