```python
from bs4 import BeautifulSoup
import requests
import urllib.request
import csv
from urllib.request import urlopen
from datetime import datetime
import time
import pandas as pd
import os
import math
from selenium import webdriver
from selenium.webdriver.common.keys import Keys

with open ('F:\\SKRIPSI\\Topik2\\UjianValidasi\\LinkURL\\url_gramedia3.csv') as f:
    
    urls=(line.strip()for line in f)
    gramedia_list= []
    for row in urls:
        print(row)
        driver = webdriver.Firefox()
        driver.get(row)
        soup = BeautifulSoup(driver.page_source, "html.parser")
        
        try:
            page= soup.find('gm-pagination', {'class':'ng-star-inserted'})
            page= page.find_all('span')[0].text
            page= page.replace('1 / ','')
            page= int(page)
            print(page)
        except:
            page= '0'
            
        if page == '0':
            all_product= soup.find_all('gm-product-list',{'class':"ng-star-inserted"})
            for item in all_product:
                try:
                    linkEcommerce = row
                except :
                    linkEcommerce = ''
                try:
                    linkProduk= item.find_all('a', {'class':"display-blocks"})[0]
                    linkProduk= linkProduk.get('href')
                except:
                    linkProduk= ''
                try:
                    namaProduk= item.find_all('p',{'class':'product-name'})[0].text
                except:
                    namaProduk= ''
                try:
                    hargaLama= item.find_all('p',{'class':'ribbon-image-price-normal'})[0].text
                except:
                    hargaLama=''
                try:
                    author= item.find_all('p',{'class':'product-authors ng-star-inserted'})[0].text
                except:
                    author=''
                gramedia=[linkEcommerce,namaProduk, author, linkProduk ,hargaLama , datetime.now().strftime("%y-%m-%d %H:%M")]
                gramedia_list.append(gramedia)
                
                
        else:
            
            for i in range(page):
                change_url= row.replace(row, row+'&page=%s'%(i+1))
                print(change_url)
                driver = webdriver.Firefox()
                driver.get(change_url)
                next_soup = BeautifulSoup(driver.page_source, "html.parser")
                all_product= next_soup.find_all('gm-product-list',{'class':"ng-star-inserted"})
                for item in all_product:
                    try:
                        linkEcommerce = row
                    except :
                        linkEcommerce = ''
                    try:
                        linkProduk= item.find_all ('a', {'class':"display-blocks"})[0]
                        linkProduk= linkProduk.get('href')
                    except:
                        linkProduk= ''
                    try:
                        namaProduk= item.find_all('p',{'class':'product-name'})[0].text
                    except:
                        namaProduk= ''
                    try:
                        hargaLama= item.find_all('p',{'class':'ribbon-image-price-normal'})[0].text
                    except:
                        hargaLama=''
                    try:
                        author= item.find_all('p',{'class':'product-authors ng-star-inserted'})[0].text
                    except:
                        author=''
                    gramedia=[linkEcommerce,namaProduk, author, linkProduk ,hargaLama , datetime.now().strftime("%y-%m-%d %H:%M")]
                    gramedia_list.append(gramedia)
                    
            
    path = 'F:\\SKRIPSI\\Topik2\\UjianValidasi\\HasilScraper\\'
    now = datetime.now()
    filename = 'gramedia2_%s_%s_%s %s_%s_%s_%s.csv'%(now.day,now.month,now.year,now.hour,now.minute,now.second,now.microsecond)
    df=pd.DataFrame(gramedia_list, columns=['linkEcommerce','namaProduk', 'author', 'linkProduk' ,'hargaLama' , 'tgl update'])
    df.to_csv(os.path.join(path, filename))  
```
